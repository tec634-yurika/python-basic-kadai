```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数numに0～30までのランダムな整数を代入する
num = random.randint(0, 30)

if num % 3 == 0 and num % 5 == 0:   # 変数numが3と5の倍数の場合
    print("3と5の倍数です")
elif num % 3 == 0:          # 変数numが3の倍数の場合
    print("3の倍数です")
elif num % 5 == 0:          # 変数numが5の倍数の場合
    print("5の倍数です")
else:                       # それ以外の場合
    print(num)
```
