```python
class Human:
    def __init__(self, name, age):
      # 属性を定義する
      self.name = name
      self.age = age

    # 大人かどうかを判定して標準出力するメソッド
    def check_adult(self):
      if self.age >= 20:
          print(f"{self.name} {self.age}才【大人です】")
      else:
          print(f"{self.name} {self.age}才【大人ではないです】")

# インスタンスを作成する
human1 = Human("山田太郎", 28)
human2 = Human("山田一郎", 12)
human3 = Human("山田二郎", 38)
human4 = Human("山田三郎", 20)
human5 = Human("山田花子", 18)

# インスタンスをリストに追加する
human_list = [human1, human2, human3, human4, human5]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in human_list:
    human.check_adult()
```
